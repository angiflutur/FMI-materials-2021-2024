```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
```

```python
image0 = np.load("car_0.npy")
image1 = np.load("car_1.npy")
image2 = np.load("car_2.npy")
image3 = np.load("car_3.npy")
image4 = np.load("car_4.npy")
image5 = np.load("car_5.npy")
image6 = np.load("car_6.npy")
image7 = np.load("car_7.npy")
image8 = np.load("car_8.npy")
```

```python
# a. Citiți imaginile din aceste fișiere și salvați-le într-un np.array
# (va avea dimensiunea 9x400x600).

a = np.array([image0, image1, image2, image3, image4, image5, image6, image7, image8])
```

```python
# b. Calculați suma valorilor pixelilor tuturor imaginilor.
suma = np.sum(a)
print(suma)
```

```python
# c. Calculați suma valorilor pixelilor pentru fiecare imagine în parte.
for i in a:
    print(np.sum(i))
```

```python
# d. Afișați indexul imaginii cu suma maximă.
maxim = 0
for i in range(9):
    if np.sum(i) > maxim:
        maxim = np.sum(a[i])
        index = i
print(index)
```

```python
# e. Calculați imaginea medie și afișati-o.

# axis = 0 -> indică faptul că dorim să realizăm operația pe toate cele n imagini,
# păstrând dimensiunile pixelilor neschimbate
imag_med = a.mean(axis=0)

io.imshow(imag_med.astype(np.uint8))

io.show()
```

```python
# f. Cu ajutorul funcției np.std(images_array), calculați deviația standard a
# imaginilor. 
dev = np.std(a)
print(dev)
```

```python
# g. Normalizați imaginile. (se scade imaginea medie și se împarte rezultatul la
# deviația standard).
for i in range(9):
    aux = (a[i] - imag_med) / dev
    io.imshow(aux.astype(np.uint8))
    io.show()
```

```python
# h. Decupați fiecare imagine, afișând numai liniile cuprinse între 200 și 300,
# respectiv coloanele cuprinse între 280 și 400.
for i in a:
    aux = i[200:300, 280:400]
    io.imshow(aux.astype(np.uint8))
    io.show()
```
